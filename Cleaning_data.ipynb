```python
def tag_checker(my_type):
    all_tags = set()
    for root, dirs, files in os.walk(".", topdown=True):
        if len(files) > 0:
            for file in files:
                if file.split('.')[-1] == "not":
                    continue
                if file.split('-')[1].split('_')[1] == my_type:
                    with open(os.path.join(root, file), 'r', encoding='utf-8') as f:
                        tmp_str = ''
                        for line in f:
                            tmp_str = tmp_str + ' ' + line.strip()
                        tmp_str = tmp_str.split('<body>')[1].split('</body>')[0].strip()
                        all_tags = all_tags | set(re.findall(r'<\w+>|</\w+>|<\w+/>|<[\w|"]*>', tmp_str))
                        if my_type == 'per':
                            all_tags = all_tags.difference({'<persName>', '</persName>', '<persName/>'})
                        elif my_type == 'org':
                            all_tags = all_tags.difference({'<orgName>', '</orgName>', '<orgName/>'})
                        elif my_type == 'loc':
                            all_tags = all_tags.difference({'<placeName>', '</placeName>'})
                        elif my_type == 'mea':
                            all_tags = all_tags.difference({'<[\w|"]*>', '</measure>'})
                            #all_tags = set([tag for ])
                        elif my_type == 'oth':
                            all_tags = all_tags.difference({'<name [\w|"]*>', '</name>', '<name>'})
                        elif my_type == 'dat':
                            all_tags = all_tags.difference({'<date [\w|"]*>', '</date>', '<date>'})
                        else:
                            print(my_type)
                            print('wtf')
                            
                        #for i
                        #print(tmp_str)
                        #print('---------------------------------------------------')
    return list(all_tags)

def make_rule(rule_list):
    final_rule = ''
    for rule in range (len(rule_list)-1):
        final_rule = final_rule + rule_list[rule] + '|'
    final_rule = final_rule + rule_list[-1]
    return final_rule

def cleaning_type(my_type, not_use_tags):
    cnt_sentence = 0
    with open('..\\data\\' + my_type+'.txt', 'w', encoding = 'utf8') as data_file:
        for root, dirs, files in os.walk(".", topdown=True):
            if len(files) > 0:
                for file in files:
                    if file.split('.')[-1] == "not":
                        continue
                    if file.split('-')[1].split('_')[1] == my_type:
                        with open(os.path.join(root, file), 'r', encoding='utf-8') as f:
                            tmp_str = ''
                            for line in f:
                                tmp_str = tmp_str + ' ' + line.strip()
                            tmp_str = tmp_str.split('<body>')[1].split('</body>')[0].strip()
                            tmp_str = re.sub(not_use_tags, '  ', tmp_str)
                            cnt_sentence = cnt_sentence + 1
                            #data_file.write(' '.join(tmp_str.strip().split()) + '\n')
                            tmp_str = ' '.join(tmp_str.strip().split())
                            tmp_str = re.split(r'<[a-zA-Z0-9ก-๙=\"\/ \.]*>', tmp_str.strip())
                            data_file.write('||'.join(tmp_str) + '\n')
                            data_file.write('-----------------------------------------------------' + '\n')
    return cnt_sentence
```

```python
#my_type = 'org'
all_type = ['mea', 'org', 'dat', 'loc', 'per', 'oth']
with open('..\\data\\report.txt', 'w', encoding = 'utf8') as F:
    for my_type in all_type:
        All_tags = tag_checker(my_type)
        cnt_sentence = cleaning_type(my_type, make_rule(All_tags))
        F.write(f'Total {my_type} sentence: {cnt_sentence}\n')
        print(my_type)
```
